# 1. Демонстрация основных методов Pandas

Pandas - это мощная библиотека Python, которая позволяет легко анализировать данные. Она особенно полезна для работы с данными, хранящимися в табличных форматах, таких как .csv, .tsv или .xlsx. С помощью Pandas вы можете легко загружать, обрабатывать и анализировать данные, используя SQL-подобные команды. При использовании в сочетании с Matplotlib и Seaborn Pandas предоставляет широкие возможности для визуализации и анализа табличных данных.

Основными структурами данных в Pandas являются Series и DataFrames. Серия - это одномерный индексированный массив данных одного типа, а DataFrame - это двумерная таблица, в которой каждый столбец содержит данные одного типа. Считайте, что DataFrame - это коллекция объектов Series. DataFrame идеально подходят для представления реальных данных, каждая строка которых представляет собой экземпляр.

## Индексирование и извлечение данных

DataFrame можно индексировать несколькими различными способами.

Чтобы получить один столбец, можно использовать конструкцию DataFrame['Name']. Давайте воспользуемся ею, чтобы ответить на вопрос, касающийся только этого столбца: какова доля откликнувшихся пользователей в нашем фрейме данных?

````python
df["Churn"].mean()
````

14,5 % - это очень плохо для компании; такой показатель оттока может привести ее к банкротству.

Булевая индексация с одним столбцом также очень удобна. Синтаксис - df[P(df['Name'])], где P - некоторое логическое условие, которое проверяется для каждого элемента столбца Name. Результатом такой индексации является DataFrame, состоящий только из строк, удовлетворяющих условию P для столбца Name.

Используем его для ответа на вопрос:

Каковы средние значения числовых характеристик для отторгаемых пользователей?

Здесь мы прибегнем к дополнительному методу select_dtypes, чтобы выбрать все числовые столбцы.

````python
df.select_dtypes(include=np.number)[df["Churn"] == 1].mean()
````

Сколько времени (в среднем) ушедшие пользователи тратят на телефон в течение дня?

````python
df[df["Churn"] == 1]["Total day minutes"].mean()
````

Какова максимальная продолжительность международных звонков среди лояльных пользователей (Chrun == 0), у которых нет международного тарифного плана?

````python
df[
    (df["Churn"] == 0) & (df["International plan"] == "No")
]["Total intl minutes"].max()
````

DataFrames можно индексировать по имени столбца (метка), имени строки (индекс) или по порядковому номеру строки. Для индексации по имени используется метод loc, а для индексации по номеру - iloc().

В первом случае ниже мы говорим: «Дайте нам значения строк с индексом от 0 до 5 (включительно) и столбцов с метками от State до Area code (включительно)». Во втором случае мы говорим «дайте нам значения первых пяти строк в первых трех столбцах» (как в типичном срезе Python: максимальное значение не включается).

````python
df.loc[0:5, "State":"Area code"]
````

````python
df.iloc[0:5, 0:3]
````

Если нам нужна первая или последняя строка фрейма данных, мы можем использовать конструкцию df[:1] или df[-1:]:

````python
df[-1:]
````

## Группировка

В общем случае группировка данных в Pandas работает следующим образом:

```python
df.groupby(by=grouping_columns)[columns_to_show].function()
```

1. Сначала метод groupby разделяет столбцы grouping_columns по их значениям. Они становятся новым индексом в результирующем кадре данных.

2. Затем выбираются интересующие столбцы (columns_to_show). Если columns_to_show не включен, будут включены все столбцы, не относящиеся к groupby.

3. Наконец, к полученным группам по выбранным столбцам применяется одна или несколько функций.

Приведем пример, в котором мы группируем данные по значениям переменной Churn и выводим статистику по трем столбцам в каждой группе:

````python
columns_to_show = ["Total day minutes", "Total eve minutes", "Total night minutes"]

df.groupby(["Churn"])[columns_to_show].describe(percentiles=[])
````

## Сводные таблицы

Предположим, мы хотим посмотреть, как распределяются наблюдения в нашем наборе данных в контексте двух переменных - `Churn` и `International plan`. Для этого мы можем построить таблицу случайностей, используя метод `crosstab`:

````python
pd.crosstab(df["Churn"], df["International plan"])
````

````python
pd.crosstab(df["Churn"], df["Voice mail plan"], normalize=True)
````

Мы видим, что большинство пользователей лояльны и не пользуются дополнительными услугами (международный план/голосовая почта).

Для тех, кто знаком с Excel, это будет напоминать поворотные таблицы. И, конечно же, таблицы pivot реализованы в Pandas: метод `pivot_table` принимает следующие параметры:

- `values` - список переменных для расчета статистики,

- `index` - список переменных, по которым нужно сгруппировать данные,

- `aggfunc` - какая статистика должна быть вычислена для групп, например, сумма, среднее, максимум, минимум или что-то еще.

Давайте посмотрим на среднее количество дневных, вечерних и ночных звонков по коду города:

````python
df.pivot_table(
    ["Total day calls", "Total eve calls", "Total night calls"],
    ["Area code"],
    aggfunc="mean",
)
````

## Преобразования DataFrame

Как и многое другое в Pandas, добавление столбцов в `DataFrame` может быть выполнено различными способами.

Например, если мы хотим подсчитать общее количество звонков для всех пользователей, давайте создадим серию `total_calls` и вставим ее в DataFrame:

````python
total_calls = (
    df["Total day calls"]
    + df["Total eve calls"]
    + df["Total night calls"]
    + df["Total intl calls"]
)
df.insert(loc=len(df.columns), column="Total calls", value=total_calls)
# параметр loc — это количество столбцов, после которых вставляется объект Series 
# мы устанавливаем его в len(df.columns), чтобы вставить его в самый конец dataframe
df.head()
````

Можно проще добавить столбец, не создавая промежуточный экземпляр Series:

````python
df["Total charge"] = (
    df["Total day charge"]
    + df["Total eve charge"]
    + df["Total night charge"]
    + df["Total intl charge"]
)
df.head()
````

Для удаления столбцов или строк используйте метод `drop`, передавая необходимые индексы и параметр `axis` (`1`, если вы удаляете столбцы, и ничего или `0`, если вы удаляете строки). Аргумент `inplace` указывает, следует ли изменять исходный DataFrame. При `inplace=False` метод `drop` не изменяет существующий DataFrame и возвращает новый с удаленными строками или столбцами. При `inplace=True` он изменяет DataFrame.

````python
# избавиться от только что созданных столбцов
df.drop(["Total charge", "Total calls"], axis=1, inplace=True)
# и вот как можно удалить строки
df.drop([1, 2]).head()
````

# 2. Первая попытка предсказать отток абонентов

Давайте посмотрим, как показатель оттока абонентов связан с функцией международного плана. Для этого мы воспользуемся таблицей случайностей `crosstab`, а также проведем визуальный анализ с помощью `Seaborn`.

````python
pd.crosstab(df["Churn"], df["International plan"], margins=True)
````

````python
import matplotlib.pyplot as plt

# !pip install seaborn
import seaborn as sns

# import some nice vis settings
sns.set()
# Graphics in the Retina format are more sharp and legible
%config InlineBackend.figure_format = 'retina'
````

````python
sns.countplot(x="International plan", hue="Churn", data=df);
````

Мы видим, что процент оттока абонентов значительно выше при использовании Международного плана. Это заслуживает внимания. Возможно, высокие и плохо управляемые расходы на международные звонки вызывают конфликты и приводят к недовольству клиентов оператора связи.

Далее рассмотрим еще одну важную характеристику - звонки в службу поддержки. Также составим сводную таблицу и картинку.

````python
pd.crosstab(df["Churn"], df["Customer service calls"], margins=True)
````

````python
sns.countplot(x="Customer service calls", hue="Churn", data=df);
````

Хотя из сводной таблицы это не так заметно, из приведенного выше графика легко понять, что процент оттока резко возрастает с 4 обращений в службу поддержки и выше.

Теперь давайте добавим в наш DataFrame бинарную характеристику - `Customer service calls > 3`. И снова посмотрим, как она связана с оттоком.

````python
df["Many_service_calls"] = (df["Customer service calls"] > 3).astype("int")

pd.crosstab(df["Many_service_calls"], df["Churn"], margins=True)
````

````python
sns.countplot(x="Many_service_calls", hue="Churn", data=df);
````

Давайте построим еще одну таблицу непредвиденных обстоятельств, которая связывает Churn как с международным планом, так и с только что созданной функцией Many_service_calls.

````python
pd.crosstab(
    df["Many_service_calls"] & df["International plan"], 
    df["Churn"], 
    margins=True
)
````

Таким образом, предсказывая, что клиент не останется лояльным (Churn=1), если он совершил более 3 обращений в сервисный центр И добавил международный план, и предсказывая Churn=0 в противном случае (а «в противном случае» здесь означает отрицание, то есть Many_service_calls <= 3 ИЛИ международный план не добавлен), мы ожидаем точность 85,8% (мы ошибемся всего 464 + 9 раз, посмотрите на таблицу случайностей выше; и $1 - \frac{464+9}{3333} \approx 85.8 \%$). Точность 85,8%, полученная в результате таких простых рассуждений, служит полезной отправной точкой (baseline) для разработки будущих моделей машинного обучения.

По мере изучения этого курса вспомните, что до появления машинного обучения процесс анализа данных выглядел примерно так. Давайте вспомним, что мы уже изучили:

- Доля лояльных клиентов в наборе данных составляет 85,5 %. Самая наивная модель, которая всегда предсказывает «лояльного клиента» на таких данных, угадает правильно примерно в 85,5 % случаев. То есть доля правильных ответов (точность) последующих моделей должна быть не меньше этого числа, а надеемся, и значительно выше;

- С помощью простого предсказания, которое можно выразить следующей формулой: `International plan = True & Customer Service calls > 3 => Churn = 1, else Churn = 0`, мы можем ожидать, что процент угадывания составит 85,8%, что чуть выше 85,5%.

- Мы получили эти две базовые модели без применения машинного обучения, и они послужат отправной точкой для наших последующих моделей.

- Прежде чем обучать сложные модели, рекомендуется немного поработать с данными, построить графики и проверить простые предположения. Более того, в бизнес-приложениях машинного обучения обычно начинают с простых решений, а затем экспериментируют с более сложными.
